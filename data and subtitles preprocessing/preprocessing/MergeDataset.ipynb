```python
import pandas as pd
```

```python
df1 = pd.read_csv('first_1000_imdbs.csv')
df1 = df1.drop('Unnamed: 0', axis=1)
df1
```

```python
df2 = pd.read_csv('second_1000_imdbs.csv')
df2 = df2.drop('Unnamed: 0', axis=1)
df2
```

```python
df3 = pd.read_csv('third_1000_imdbs.csv')
df3 = df3.drop('Unnamed: 0', axis=1)
df3
```

```python
df4 = pd.read_csv('fourth_1000_imdbs.csv')
df4 = df4.drop('Unnamed: 0', axis=1)
df4
```

```python
df = pd.concat([df1, df2, df3, df4], axis=0)
df
```

```python
df = df[df.imdbRating != '(-1, -1)']
df
```

```python
df.reset_index(drop=True, inplace=True)
```
