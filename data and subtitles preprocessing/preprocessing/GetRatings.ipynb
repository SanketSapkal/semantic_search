```python
# Vasudha's API key: 68dd8045
# Chaitanya's API key: dc32211d
# Divya's API key: 1c4964e2, f2d69a7e

import requests

def getImdbRating(moviename):
    moviename = moviename.lower()
    url = f"https://www.omdbapi.com/?t={moviename}&apikey=f2d69a7e"
    response = requests.get(url)
    data = response.json()
    if 'Response' in data and data['Response'] == 'False':
        return (-1, -1)
    return data["imdbRating"]
```

```python
df1_first1000['imdbRating'] = df1_first1000['Title'].apply(lambda func: getImdbRating(func))
```
