```python
import pandas as pd
```

```python
df = pd.read_csv('imdb_ratings.csv')
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df.loc[df['Title'] == 'Keane']
```

```python
sorted_df = df.sort_values(by=['imdbRating'], ascending=False)
sorted_df
```

```python
df1 = sorted_df.head(50)
df1
```

```python
df2 = sorted_df.loc[sorted_df['imdbRating'] <= 8.0]
df2 = df2.head(770)
df2 = df2.sample(n=30)
df2
```

```python
df3 = sorted_df.loc[sorted_df['imdbRating'] < 7.0]
df3 = df3.head(1376)
df3 = df3.sample(n=25)
df3
```

```python
df4 = sorted_df.loc[sorted_df['imdbRating'] < 6.0]
df4 = df4.head(916)
df4 = df4.sample(n=25)
df4
```

```python
df5 = sorted_df.loc[sorted_df['imdbRating'] < 5.0]
df5 = df5.head(267)
df5 = df5.sample(n=20)
df5
```

```python
df_for_subs = pd.concat([df1, df2, df3, df4, df5], axis=0)
df_for_subs = df_for_subs.sample(frac=1).reset_index(drop=True)
df_for_subs
```

```python
df_for_subs.to_csv('100_imdb_ratings_2.csv')
```
