```python
from bs4 import BeautifulSoup
def data_cleaning(line):
#     clean_text = line.lower() #lowercase all alphabets
#     clean_text = contractions.fix(clean_text) #expand contractions
    clean_text = BeautifulSoup(line).get_text() #remove html tags
    clean_text = re.sub(r'https?://\S+', '', clean_text) #remove url
#     clean_text = re.sub(r'[^\w\s]', '', clean_text) #remove punctuation
#     clean_text = re.sub(r'\d+', '', clean_text) #remove numbers
    clean_text = re.sub('\s{2,}', ' ', clean_text) #remove extra spaces
    return clean_text
```

```python
df3['clean_subs1'] = df3['Subtitles'].apply(lambda func: data_cleaning(func))
```

```python
def clean_more(line):
    if 'Subtitles by' in line:
        line = line.replace('Subtitles by', '')
    if 'Resync by' in line:
        line = line.replace('Resync by', '')
    if 'downloaded by' in line:
        line = line.replace('downloaded by', '')
#     clean_text = re.sub(r'[^\w\d\s\'\",.!?:;-]', '', line) #remove punctuation
#     "[^\w\d\s\'\",.!?:;-]+"g
    return line
```

```python
df3['clean_subs1'] = df3['clean_subs1'].apply(lambda func: clean_more(func))
```
